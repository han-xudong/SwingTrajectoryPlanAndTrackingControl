# Results Analysis

## Swing with Different Foot Step Displacement

### Unoptimized swing path

```python
from matplotlib import pyplot as plt
import numpy as np

_FOOT_STEP_DISP_LIST = np.array([0.1, 0.2, 0.3])
_WITH_OPTIMIZATION = False

# Load data
foot_path_list = []
foot_opt_path_list = []
foot_real_path_list = []
for foot_step_disp in _FOOT_STEP_DISP_LIST:
        parameters = "_disp_%.2f"%foot_step_disp + \
                      ("_optimized" if _WITH_OPTIMIZATION else "")
        path = "results/my_swing_example" + parameters + "/"
        foot_path_list.append(np.loadtxt('%sfoot_path%s.txt'%(path, parameters), delimiter=','))
        if _WITH_OPTIMIZATION:
                foot_opt_path_list.append(np.loadtxt('%sfoot_opt_path%s.txt'%(path, parameters), delimiter=','))
        foot_real_path_list.append(np.loadtxt('%sfoot_real_path%s.txt'%(path, parameters), delimiter=','))

# Plot
for figNum in range(len(foot_path_list)):
    foot_path = foot_path_list[figNum]
    if _WITH_OPTIMIZATION:
        foot_opt_path = foot_opt_path_list[figNum]
    plt.figure(figNum)
    ax = plt.axes(projection='3d')
    ax.plot3D(foot_path[:, 0], foot_path[:, 1], foot_path[:, 2], 'b')
    if _WITH_OPTIMIZATION:
        ax.plot3D(foot_opt_path[1:, 0], foot_opt_path[1:, 1], foot_opt_path[1:, 2], 'r')
    ax.set_xlabel('x (mm)')
    ax.set_ylabel('y (mm)')
    ax.set_zlabel('z (mm)')
    ax.set_xlim([0, 0.3])
    ax.set_ylim([-0.1, 0.1])
    ax.set_zlim([-0.3, -0.2])
    ax.set_xticks([0, 0.05, 0.1, 0.15, 0.2, 0.25, 0.3])
    ax.set_yticks([-1.0, -0.5, 0, 0.5, 1.0])
    ax.set_zticks([-0.3, -0.28, -0.26, -0.24, -0.22, -0.20])
    if _WITH_OPTIMIZATION:
        plt.savefig('results/my_swing_example_disp_%.2f_optimized/path.png'%_FOOT_STEP_DISP_LIST[figNum], dpi=300)
    else:
        plt.savefig('results/my_swing_example_disp_%.2f/path.png'%_FOOT_STEP_DISP_LIST[figNum], dpi=300)
```
